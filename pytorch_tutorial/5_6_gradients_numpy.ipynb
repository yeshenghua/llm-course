```python
# mannually implement a simple gradient desent example using numpy
import numpy as np

# f = w * x = 2 * x
X = np.array([1,2,3,4], dtype=np.float32)
Y = np.array([2,4,6,8], dtype=np.float32)

# parameter initialization
w = 0.0

# model prediction
def forward(x):
    return w * x

# loss = MSE
def loss(y, y_predicted):
    return ((y - y_predicted) ** 2).mean()

# gradient
# MSE = 1/N * (w*x - y) ** 2
# dJ/dw = 1/N * 2* (w*x - y)
def gradient(x,y,y_predicted):
    return np.dot(2 * x, y_predicted - y).mean()

print(f"prediction before training: f(5) = {forward(5):.3f}")

# training
learning_rate = 0.01
n_iters = 10

for epoch in range(n_iters):
    # prediction = forward pass
    y_pred = forward(X)
    
    # loss
    l = loss(Y, y_pred)
    
    # gradients
    dw = gradient(X,Y,y_pred)
    
    # update weights
    w -= learning_rate * dw
    
    if epoch % 1 == 0:
        print(f'epoch {epoch + 1}, w = {w:.3f}, loss={l:.8f}')
        
print(f'Prediction after training: f(5) = {forward(5):.3f}')
```

```python
# replace torch with numpy operations
# mannually implement a simple gradient desent example using numpy
import torch

# f = w * x = 2 * x
X = torch.tensor([1,2,3,4], dtype=torch.float32)
Y = torch.tensor([2,4,6,8], dtype=torch.float32)

# parameter initialization
w = torch.tensor(0.0, dtype=torch.float32, requires_grad=True) 

# model prediction
def forward(x):
    return w * x

# loss = MSE
def loss(y, y_predicted):
    return ((y - y_predicted) ** 2).mean()


print(f"prediction before training: f(5) = {forward(5):.3f}")

# training
learning_rate = 0.01
n_iters = 10

for epoch in range(n_iters):
    # prediction = forward pass
    y_pred = forward(X)
    
    # loss
    l = loss(Y, y_pred)
    
    # gradients
    l.backward()
    
    # update weights, not part of the gradient tracking graph
    with torch.no_grad():
        w -= learning_rate * w.grad
        
    w.grad.zero_()
    
    if epoch % 1 == 0:
        print(f'epoch {epoch + 1}, w = {w:.3f}, loss={l:.8f}')
        
print(f'Prediction after training: f(5) = {forward(5):.3f}')
```

```python
# full version
import torch
import torch.nn as nn

# f = w * x = 2 * x
X = torch.tensor([[1],[2],[3],[4]], dtype=torch.float32)
Y = torch.tensor([[2],[4],[6],[8]], dtype=torch.float32)

# parameter initialization
w = torch.tensor(0.0, dtype=torch.float32, requires_grad=True) 

X_test = torch.tensor([[5]], dtype=torch.float32)
n_samples, n_features = X.shape
print(n_samples, n_features)

input_size = n_features
output_size = n_features

# model = nn.Linear(input_size, output_size)         # X must be an 2D array now, simple model
# do exactly the same thing
class LinearRegression(nn.Module):
    def __init__(self, input_dim, output_dim):
        super(LinearRegression, self).__init__()
        
        # define layer
        self.lin = nn.Linear(input_dim, output_dim)
    
    def forward(self, x):
        return self.lin(x)
    
    
model = LinearRegression(input_size, output_size)

print(f"prediction before training: f(5) = {model(X_test).item():.3f}")

# training
learning_rate = 0.01
n_iters = 10

loss = nn.MSELoss()             # callable function
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)

for epoch in range(n_iters):
    # prediction = forward pass
    y_pred = model(X)
    
    # loss
    l = loss(Y, y_pred)
    
    # gradients
    l.backward()
    
    # update weights, not part of the gradient tracking graph
    with torch.no_grad():
        optimizer.step()            # do a optimization step
        
    optimizer.zero_grad()
    
    if epoch % 1 == 0:
        [w, b] = model.parameters()         # unpack
        print(f'epoch {epoch + 1}, w = {w[0][0].item():.3f}, loss={l:.8f}')
        
print(f'Prediction after training: f(5) = {model(X_test).item():.3f}')
```
